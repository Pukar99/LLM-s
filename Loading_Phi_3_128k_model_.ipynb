```python
#@title Download Phi-3-mini-4k-instruct model
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-128k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-128k-instruct")
```

```python
# Install BeautifulSoup if needed
!pip install beautifulsoup4

# Import libraries
import requests
from bs4 import BeautifulSoup
```

```python

# URL of the page to scrape
url = 'https://www.arizona.edu/'

# Fetch the page
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Extract text content
text_content = soup.get_text()

# Remove extra spaces, newlines, and tabs
cleaned_text = ' '.join(text_content.split())

# Print the cleaned text content
# print(text_content)
print(cleaned_text)
cleaned_text = cleaned_text[:5000]

```

```python
# Define the messages
messages = [
    {
        "role": "user",
        "content": "You are a university page classifier to detect course, professor and other pages. Your task is to classify the text given into categories like Home Page, Course, Professor Page"
    },
    {
        "role": "assistant",
        "content": "{ 'response': [{}]}"
    },
    {
        "role": "user",
        "content": f"Classify the professor names, courses, contacts, and emails from the provided cleaned HTML text: {cleaned_text}"
    }
]

# Create the text generation pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

# Define generation arguments
generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

# Generate the response
output = pipe(messages, **generation_args)
print(output[0]['generated_text'])
```
